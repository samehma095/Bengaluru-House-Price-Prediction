```python
df1=df.dropna()
```

```python
df1['bhk'] = df1['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df1.head()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True

# هيعرض لك بس الصفوف اللي المساحة فيها مش رقم عادي (زي الـ Range)
df1[~df1['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    try:
        return float(x)
    except:
        return None

# دلوقتي هنطبق الوظيفة دي على العمود كله في نسخة جديدة من البيانات
df3 = df1.copy()
df3['total_sqft'] = df3['total_sqft'].apply(convert_sqft_to_num)
```

```python
df3.loc[30],df3.loc[122]
```

```python
df4 = df3.copy()
df4['price_per_sqft'] = df4['price'] * 100000 / df4['total_sqft']
df4.head()
```

```python
# 1. تنظيف المسافات
df4.location = df4.location.apply(lambda x: x.strip())

# 2. معرفة عدد البيوت في كل منطقة
location_stats = df4.groupby('location')['location'].agg('count').sort_values(ascending=False)

# 3. تحديد المناطق اللي تكرارها أقل من 10
locations_less_than_10 = location_stats[location_stats <= 10]

# 4. تحويل المناطق دي لـ "other"
df4.location = df4.location.apply(lambda x: 'other' if x in locations_less_than_10 else x)

# نشوف بقوا كام منطقة دلوقتي
len(df4.location.unique())
```

```python
# بنشيل أي شقة مساحتها الكلية مقسومة على عدد الأوض (bhk) أقل من 300
df5 = df4[~(df4.total_sqft/df4.bhk < 300)]
df5.shape
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft) # المتوسط
        st = np.std(subdf.price_per_sqft) # الانحراف المعياري
        # بنحتفظ بس بالبيانات اللي قريبة من المتوسط
        reduced_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df6 = remove_pps_outliers(df5)
df6.shape
```

```python
# بنشيل أي شقة فيها حمامات أكتر من (عدد الأوض + 2)
df7 = df6[df6.bath < df6.bhk + 2]
df7.shape
```

```python
# حذف الأعمدة اللي مش محتاجينها
df8 = df7.drop(['size', 'price_per_sqft', 'society', 'balcony', 'availability', 'area_type'], axis='columns')
df8.head()
```

```python
# 1. تحويل المناطق لأعمدة أرقام
dummies = pd.get_dummies(df8.location)

# 2. دمج الأعمدة الجديدة مع الجدول الأصلي وحذف عمود الـ "other" (عشان نتجنب مشكلة Dummy Variable Trap)
df9 = pd.concat([df8, dummies.drop('other', axis='columns')], axis='columns')

# 3. حذف عمود الـ location الأصلي لأننا خلاص حولناه لأرقام
df10 = df9.drop('location', axis='columns')

df10.head()
```

```python
X = df10.drop('price', axis='columns') # كل الأعمدة ماعدا السعر
y = df10.price # عمود السعر فقط
```

```python
from sklearn.model_selection import train_test_split

# بنقسم البيانات: 80% تدريب و 20% اختبار
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
     from sklearn.linear_model import LinearRegression

lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train) # هنا الموديل "بيذاكر" البيانات

# قياس الدقة
lr_clf.score(X_test, y_test)
```

```python
def predict_price(location, sqft, bath, bhk):    
    loc_index = np.where(X.columns == location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1

    return lr_clf.predict([x])[0]
```

```python
predict_price('1st Phase JP Nagar', 1000, 2, 2)
```

```python
predict_price('Indira Nagar', 1000, 2, 2)
```
